```python
import polars as pl

orders_df = pl.DataFrame({
    "item": ["pen", "cup", "pen", "fork"],
    "price": [1.5, None, 1.5, 2.2],
    "qty": [1, 1, 2, 4],
})

orders = skrub.var("orders", orders_df)
orders
```

```python
orders = orders.group_by("item").agg(pl.sum("price"))
```

```python
orders = orders.select(pl.sum("price"))
```

```python
print(orders.skb.describe_steps())
```
